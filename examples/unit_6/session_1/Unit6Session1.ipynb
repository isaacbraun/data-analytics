## Data splitting

```python
#From: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

import numpy as np
from sklearn.model_selection import train_test_split
X, y = np.arange(10).reshape((5, 2)), range(5)
print(X)
print(list(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

## Model evaluation

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report 
y_true = [0, 1, 2, 2, 2]
y_pred = [0, 0, 2, 2, 1]
target_names = ['class 0', 'class 1', 'class 2'] 
confusion_matrix(y_true, y_pred)
print(classification_report(y_true, y_pred, target_names=target_names))

#Precision: How often a classifier is right when it says something is fraud?
#Recall: How much of the actual fraud we correctly detect


#Precision example for class 0: true positives / (true positives + false positives) 
#                               = 1 / (1 + 1)  = 0.5

#Recall example for class 0: true positives / (true positives + false negatives)
#                            = 1 / (1 + 0) = 1    
    
#Accuracy: 3 correct classifications / 5 classifications = 0.6
#Macro avg: e.g. f1-score = (0.67 + 0.00 + 0.80)/3 = 0.49
#Weighted avg: e.g. f1-score = (0.67 * (0.2... just 1 sample is from class 0 in y_true) 
#                              + 0.00 * (0.2) + 0.80 * (0.6)) = 0.61 
```
