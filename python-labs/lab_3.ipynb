# Lab 3 - KNN

## Setup

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report 
```

## Load and Look at Data

```python
iris = datasets.load_iris()
print(iris)
```

## Get X, Y, Names from Dataset

```python
x = iris.data
y = iris.target
names_array = iris.target_names
print(x)
print(y)
print(names_array)
```

## Split Data into Train/Test (80/20)

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=81)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

## Fit KNN Alogorithm (K = 7)

```python
knn_7 = KNeighborsClassifier(n_neighbors=7)
knn_7.fit(x_train, y_train)
```

## Run Predictions on Testing Set

```python
print(y_test)
print(names_array)
```

```python
y_test_pred = knn_7.predict(x_test)
print(y_test_pred)
```

## Evaluate Model

```python
confusion_matrix(y_test, y_test_pred)
```

```python
report = classification_report(y_test, y_test_pred, target_names=names_array)
print(report)
```

## Trying with fewer neighbors (K = 3)

```python
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_3.fit(x_train, y_train)

y_test_pred_3 = knn_3.predict(x_test)
print(y_test_pred_3)
```

```python
confusion_matrix(y_test, y_test_pred_3)
```

```python
report_3 = classification_report(y_test, y_test_pred, target_names=names_array)
print(report_3)
```
